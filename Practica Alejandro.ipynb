* Verificar que el signo de la función en $a$ y $b$ es distinto.
* Calcular el punto $c$ como el valor medio del intervalo $[a, b]$. Se evalua $f(c)$ y:
    * Si $f(c)$ es cero hemos encontrado una raíz (solución de la ecuación).

* Con este nuevo intervalo se vuelve a iniciar el proceso.
* El proceso continua hasta alcanzar la precisión especificada, calculada como la diferencia entre $b-a$. Es decir, si $|b - a| < precision$.
